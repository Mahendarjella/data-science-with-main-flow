```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame data
data = {
    'Month': ['January', 'February', 'March', 'April', 'May'],
    'Sales': [300, 400, 350, 500, 600],
    'Profit': [70, 90, 80, 120, 150]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting Bar Chart for Sales
plt.figure(figsize=(10, 5))
plt.bar(df['Month'], df['Sales'], color='skyblue', label='Sales')

# Customizations for Bar Chart
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales Amount ($)')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Plotting Line Chart for Profit
plt.figure(figsize=(10, 5))
plt.plot(df['Month'], df['Profit'], color='green', marker='o', label='Profit')

# Customizations for Line Chart
plt.title('Monthly Profit')
plt.xlabel('Month')
plt.ylabel('Profit Amount ($)')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
