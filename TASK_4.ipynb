```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load your dataset (replace 'your_dataset.csv' with your file path)
data = pd.read_csv('C:\\Users\\KGRCET\\Downloads\\covid_19.csv')

# Display the first few rows of the dataset
print("Dataset Head:")
print(data.head())
```

```python
data.describe()
```

```python
data.info()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(x='WHO Region', y='Deaths', data=data)
plt.title('Scatter Plot of Deaths by WHO Region')
plt.xlabel('WHO Region')
plt.ylabel('Deaths')
plt.show()
```

```python
print(data.columns)
```

```python
plt.figure(figsize=(8, 5))

# Boxplot for 'Deaths' grouped by 'WHO Region'
sns.boxplot(x='WHO Region', y='Deaths', data=data, color='green')
plt.title('Distribution of Deaths by WHO Region')
plt.xlabel('WHO Region')
plt.ylabel('Deaths')
plt.xticks(rotation=45)  # Rotate x-axis labels if needed
plt.show()
```

```python
# Check for correlations between numeric variables
plt.figure(figsize=(10, 10))

# Compute the correlation matrix
correlation_matrix = data.corr()

# Generate a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
